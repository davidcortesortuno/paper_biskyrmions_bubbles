# Isolated bubbles

Notebook for the simulation of an isolated bubble in a cuboid sample:

# Imports

```python
import joommf_bubble as jb
%matplotlib inline
```

# Simulation

Here we create a simulation with given magnetic parameters and film size. The `init_state_radius` parameter given here is the radius of the onion-like profile used as initial state for the magnetisation field.

The system size is specified with the `L` and `thickness` parameters, which are given in nanometres units. These values should be multiples of 4 (because we discretise the system in cells of 4x4x4 nm^3)

```python
sim = jb.IsolatedBubble(A=20e-12, Ms=0.648, B=0.25,
                        L=800e-9, thickness=100e-9,
                        init_state_radius=60e-9
                        )
```

We can plot how the initial state looks like at the bottom surface. To plot the top surface, use `n_slice=-1` and to plot the bottom surface, use `n=0`. Available values for the slices can be seen from the `sim.z_layers` table:

```python
sim.z_layers
```

So we can use any of the integers:

```python
sim.plot_slice(n_slice=0)
```

Here we relax the system and find the type II bubble after performing an energy minimisation:

```python
sim.minimise_energy()
```

We can finally plot again the magnetisation field after relaxation:

```python
sim.plot_slice(n_slice=0)
```

And finally we can save the data. Three files will be created, one with the coordinates and three with the agnetisation components:

```python
sim.save_data(filename='type_II_bubble')
```
